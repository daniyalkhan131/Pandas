```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('ipl_matches.csv')
```

```python
myseries=df['winner'].value_counts()
```

```python
(df['team1'].value_counts() + df['team2'].value_counts()).sort_values()
#it sorts in ascending order bydefault
```

```python
(df['team1'].value_counts() + df['team2'].value_counts()).sort_values(ascending=False)
```

```python
#applying this on dataframe is like we need to give column on which basis the
#sorting will be done

df.sort_values('city',ascending=False)#this is not making any change as not inplace
#for that we do
df.sort_values('city',ascending=False,inplace=True)
```

```python
df.head()
```

```python
#for sorting data frame based on multiple columns
df.sort_values(['city','date'])
```

```python
#for city ascending order and data descending order
df.sort_values(['city','date'],ascending=[True,False])
```

```python
#drop duplicates
```

```python
df.drop_duplicates(subset=['city']).sample(5) #this will drop all rows that have duplicate
#values in city col. and first occurence will be kept
#put inplace=True for making changes at place

df.drop_duplicates(subset=['city']).shape
```

```python
df.drop_duplicates(subset=['city','season']).shape
#it is like an and operation , rows got removed where both are duplicates
```

```python
#for getting the winner of every season we need to use drop duplicates
#we know the last match of that season will decide who won season so we drop
#all but keeping the last of the season
df.drop_duplicates('season',keep='last')[['season','winner']]
```

```python
df.drop_duplicates('season',keep='last')[['season','winner']].sort_values('season')
```

```python
import pandas as pd
import numpy as np
```

```python
company=pd.read_csv('Fortune_1000.csv')
company.head(2)
```

```python
#things that are getting repeated and are categorical we do grouping on that

company.groupby('sector')#it is not accessible like this
```

```python
sectors=company.groupby('sector')
```

```python
print(len(company))
print(len(sectors))
```

```python
sectors.size() #all are in ascending order of secroe names
```

```python
#sector having largest no. of companies
sectors.size().sort_values(ascending=False)
```

```python
#it will return name of companies which occur at first position in every
#sector

sectors.first()
#sectors.last()
```

```python
#it will return dict having sector names and id of rows coming in that sector
sectors.groups

#can check
#company.iloc[54,:]
```

```python
#for getting all the data of a particular sector
sectors.get_group('Apparel')
#sectors.get_group('Apparel').shape
```

```python
#apply sum function to the num. columns of sectors
sectors.sum()
```

```python
sectors.mean()
```

```python
sectors['revenue'].mean()
```
