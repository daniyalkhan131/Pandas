```python
#creating function for that, ki batsman ka name den aur vo bataen ki sabse zada runs kis team pr mare
def runs_scored(batsman):
  vk=delivery[delivery['batsman']==batsman]
  return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
#find the most destructive death over batsman in the history of IPL
#it is calc by Strike rate which is (no. of runs/no. of balls)*100
#and min balls batsman played is 200 in over 16-20(this is death over)
#because if bastman played 1 last baal in entire ipl and score 6 then his
#strike rate will be 600 that is not correct way to calc.
```

```python
#first we remove rows of <15 over
death_over=delivery[delivery['over']>15]
death_over.sample(5)
```

```python
#for no. of balls
all_batsman=death_over.groupby('batsman')['batsman_runs'].count()
#and we need those who played >200
all_batsman[all_batsman>200]
```

```python
#we extract these names
batsman_list=all_batsman[all_batsman>200].index.tolist() #tolist convert index object to list
```

```python
#now we calc strike rate of these batsman

#we use isin() to get data of these batsman
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
balls=final.groupby('batsman')['batsman_runs'].count()
```

```python
sr=(runs/balls)*100
```

```python
sr.sort_values(ascending=False)
```
